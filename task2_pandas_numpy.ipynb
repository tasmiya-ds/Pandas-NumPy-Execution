## Create New Columns


```python
# Total score (already exists, recalculate to show logic)
df["total_score"] = df["math_score"] + df["reading_score"] + df["writing_score"]

# Average score
df["average_score"] = df["total_score"] / 3

# Performance category
df["performance"] = np.where(df["average_score"] >= 60, "Good", "Needs Improvement")
```

## Filtering Data


```python
filtered_df = df[
    (df["parental_level_of_education"] == "high school") &
    (df["average_score"] > 70
     )
]

filtered_df.head()
```

## Grouping and Aggregation


```python
grouped = df.groupby("gender")["average_score"].agg(["mean", "count"])
grouped
```
